```python
#Q1 
''''
Write a NumPy program to create a structured array from given student name, height, class and their data types. Now sort the array on height.
'''''

class Student:
    myInstances=[]
    __slots__=['_name', '_height','_clas'] #To Get only this Attributes
    def __init__(self,name="Student",height=22,cls='c'):
        self._name=name
        self._height=height
        self._clas=cls
        self.__class__.myInstances.append(self) #Append Any Object to list to reuse it 

    #Setters & Getters
   
    def getClassName(self):
        return self._clas
    def getHeight(self):
        return self._height    
    def getStudentName(self):
        return self._name   
    def setHeight(self,v):
            self._height=v          
    def setStudentName(self,v):
            if v.strip().isdigit():
                print("invalid Input")
            elif v =='':
                print("Name Is Empty")
            else :
                self._name=v
    def setClassName(self,v):
            self._clas=v

            
            


```

```python
###Vlaidator
## validate if user enter valid input
def validateName(n):
    name=input(f"Enter Your {n}  :")
    if name != '':
        if name.isalpha(): 
            print(f"{n} is Valid")
        else:       
            print("name Not Valid")
            validateName(n) 
    else :
        print(f"{n} is Empty")
        validateName(n)        
    return name  
def validateNumber(data):
    name=input(f"Enter Your {data}  :")
    if name != '':
        if name.isdigit(): 
            print(f"{data} is Valid")
        else:       
            print("name Not Valid")
            validateNumber(data) 
    else :
            print(f"{data} is Empty")
            validateNumber(data)       
    return float(name)  
            
           
```

```python
###Get Students Data

num =input("Please Enter The Number Of Student") #Number of Users
lst=[]


if num.isdigit(): #check user input
    for i in range(int(num)):
        #make list of all users
        Student.myInstances.append(Student())
        #get name
        name=validateName('Name')
        #insert into the list of all users
        Student.myInstances[i].setStudentName(name)
        #get class
        c=validateName('Class')
        #insert into the list of all users
        Student.myInstances[i].setClassName(c)
        #get height
        h=validateNumber("Height") 
        #insert into the list of all users
        Student.myInstances[i].setHeight(h)
        #convert data to tuple
        lst.append((Student.myInstances[i].getStudentName(),
                    Student.myInstances[i].getClassName(),Student.myInstances[i].getHeight(),))
        
else:
    print("Invalid Input")

Student.myInstances.clear()
print(Student.myInstances)
print(lst)

```

```python
#sort data with numpy
#
import numpy as np  
data_type = [('name', 'S15'), ('class', 'S15'), ('height', float)]
arr=np.array(lst ,dtype=data_type)
print(np.sort(arr, order='height')[::-1]) #in Descing Order
print(np.sort(arr, order='height'))#Ascending Order
```

# Q3: Implement the L1 and L2 loss functions

Exercise: Implement the numpy vectorized version of the L2 loss. There are several way of implementing the L2 loss but you may find the function np.dot() useful. As a reminder, if  =[1,2,...,] , then np.dot(x,x) =  =02 .

L2 loss is defined as


```python
def L1(yhat, y):
    """
    Arguments:
    yhat -- vector of size m (predicted labels)
    y -- vector of size m (true labels)
    
    Returns:
    loss -- the value of the L1 loss function defined above
    """
    yhat=np.array(yhat)
    y=np.array(y)   
    ### START CODE HERE ### ( 1 line of code)
    loss = 0
    for i in range(yhat.size):  #or y (same Size "M")
        loss +=np.abs(yhat[i]-y[i])
        #print(loss)  
    ### END CODE HERE ###   
    return loss
yhat = np.array([.9, 0.2, 0.1, .4, .9])

y = np.array([1, 0, 0, 1, 1])
l=L1(yhat, y)
print("L1 = ",l)
```

```python
def L2(yhat, y):

    loss=0
    for i in range(y.size):
        loss +=(y[i]-yhat[i])**2
        
    ### END CODE HERE ###
    
    return loss
```

```python
yhat = np.array([.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 1, 1])

l2_loss = L2(yhat,y)

print("L2 = " + str(l2_loss))
```
